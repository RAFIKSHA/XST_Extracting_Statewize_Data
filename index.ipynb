```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
url = "https://raw.githubusercontent.com/xscientisttech/detaset/main/india-state-wise-data-analysis.csv"
data = pd.read_csv(url)

# Display the first few rows of the dataframe
data.head()
```

```python
# Preprocess data
def preprocess_data(df):
    # Split combined data in the 'state & district' column
    data[['district_code','state_name', 'district_name']] = data['State & District'].str.split(n=2, expand=True)
    data.drop(columns=['State & District'], inplace=True)
    
    return df

# Preprocess the data
preprocessed_set= preprocess_data(data.copy())

# Displaying preprocessed data
print("Preprocessed Data is:")
preprocessed_set.head()
```

```python


# Create a bar graph for population distribution
plt.figure(figsize=(12, 6))
plt.bar(data['state_name'], data['Population'])
plt.xticks(rotation=100)
plt.xlabel('State')
plt.ylabel('Population')
plt.title('Population Distribution by Statewise')

plt.show()
```

```python
#3What is the literacy rate of each state?
#Let's create a horizontal bar graph to visualize the literacy rate across states.

plt.figure(figsize=(10, 8))
plt.barh(data['state_name'], data['Literate'])
plt.xlabel('Literacy Rate')
plt.ylabel('state')
plt.title('Literacy Rate by state')
plt.tight_layout()
plt.show()
```

```python
#Question 3: Which state has the highest and lowest GDP?
#Let's create a bar graph to compare the GDP of each state.

plt.figure(figsize=(12, 6))
plt.bar(data['state_name'], data['GDP (in Crores)'])
plt.xticks(rotation=90)
plt.xlabel('State/UT')
plt.ylabel('GDP (in Crores)')
plt.title('GDP by State/UT')
plt.tight_layout()
plt.show()

```

```python
#Question 4: What is the ratio of urban to rural population in each state?
#Let's create a stacked bar graph to compare the urban and rural populations.

plt.figure(figsize=(12, 6))
plt.bar(data['State/UT'], data['Urban Population'], label='Urban')
plt.bar(data['State/UT'], data['Rural Population'], bottom=data['Urban Population'], label='Rural')
plt.xticks(rotation=90)
plt.xlabel('State/UT')
plt.ylabel('Population')
plt.title('Urban vs Rural Population by State/UT')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
Question 5: What is the distribution of female to male ratio in each state?
Let's create a grouped bar graph to compare the female and male population ratios.

python
Copy code
plt.figure(figsize=(12, 6))
bar_width = 0.4
plt.bar(data['State/UT'], data['Female Ratio'], width=bar_width, label='Female')
plt.bar(data['State/UT'], data['Male Ratio'], width=bar_width, label='Male', bottom=data['Female Ratio'])
plt.xticks(rotation=90)
plt.xlabel('State/UT')
plt.ylabel('Population Ratio')
plt.title('Female vs Male Population Ratio by State/UT')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
Question 10: What is the correlation between GDP and Literacy Rate?
Let's create a scatter plot to visualize the relationship between GDP and Literacy Rate.

python
Copy code
plt.figure(figsize=(8, 6))
plt.scatter(data['GDP (in Crores)'], data['Literacy Rate'])
plt.xlabel('GDP (in Crores)')
plt.ylabel('Literacy Rate')
plt.title('Correlation between GDP and Literacy Rate')
plt.tight_layout()
plt.show()
```

```python
Question 9: How does the state-wise distribution of area look?
Let's create a pie chart to represent the distribution of area across states.

python
Copy code
import plotly.express as px

fig = px.pie(data, values='Area (in sq km)', names='State/UT', title='Area Distribution by State/UT')
fig.show()
```

```python
Question 8: What is the distribution of rural and urban literacy rates?
Let's create a box plot to visualize the distribution of rural and urban literacy rates.

python
Copy code
plt.figure(figsize=(8, 6))
plt.boxplot([data['Rural Literacy Rate'], data['Urban Literacy Rate']], labels=['Rural', 'Urban'])
plt.ylabel('Literacy Rate')
plt.title('Distribution of Rural and Urban Literacy Rates')
plt.tight_layout()
plt.show()
```

```python
Question 7: Which states have the highest and lowest sex ratio?
Let's create a bar graph to compare the sex ratio of each state.

python
Copy code
plt.figure(figsize=(12, 6))
plt.bar(data['State/UT'], data['Sex Ratio'])
plt.xticks(rotation=90)
plt.xlabel('State/UT')
plt.ylabel('Sex Ratio')
plt.title('Sex Ratio by State/UT')
plt.tight_layout()
plt.show()
```

```python
Question 6: What is the distribution of different religious groups in each state?
Let's create a grouped bar graph to compare the population of various religious groups.

python
Copy code
religions = ['Hindu', 'Muslim', 'Christian', 'Sikh', 'Buddhist', 'Jain', 'Other']
plt.figure(figsize=(12, 6))
for religion in religions:
    plt.bar(data['State/UT'], data[religion], label=religion)
plt.xticks(rotation=90)
plt.xlabel('State/UT')
plt.ylabel('Population')
plt.title('Religious Population by State/UT')
plt.legend()
plt.tight_layout()
plt.show()
```
